```python
fig = plt.figure()
l = np.linspace(-1,1,100)

ax1 = fig.add_subplot(1,1,1)
data1 = l**2+l-1
data2 = np.sin(l)
ax1.plot(l,data1, "g-.")
ax1.plot(l,data2, "b")
ax1.set_xlabel("$x$")
ax1.set_ylabel("$x^2+x-1$")
ax1.set_xlim(-0.75,0.75)
ax1.set_ylim(-1.5,0.5)
ax1.set_title("Plot of function $x^2+x-1$")
xticks = np.arange(-0.6, 0.7, 0.15)
ax1.set_xticks(xticks)
ax1.grid()
plt.plot()
fig.savefig("fig.pdf")
```
