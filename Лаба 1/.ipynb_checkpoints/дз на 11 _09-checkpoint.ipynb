```python
import numpy as np
import matplotlib.pyplot as plt


x1 = np.array([45.9, 50.4, 55.3, 61.4, 67.3, 75.0, 82.0, 104.5, 88.3, 77.0, 65.9, 59.6, 54.0])
y1 = np.array([100, 107.5, 117.5, 130, 145, 160, 175, 222.5, 187.5, 161, 140, 125, 115])
p1, v1 = np.polyfit(x1, y1, deg=1, cov=True)
pol1 = np.poly1d(p1)
dy1 = np.sqrt(((pol1(x1)-y1)**2/(len(x1) * (len(x1) - 1))).sum())
sx1 = x1.sum()/len(x1)
dx1 = np.sqrt(((sx1-x1)**2/(len(x1) * (len(x1) - 1))).sum())
print(dx1)
plt.errorbar(x1, y1, xerr=dx1, yerr=dy1, fmt='.', ecolor='red')
print(dy1, (pol1(x1)-y1).std() / np.sqrt(len(x1)-1))
x = np.arange(35, 110, 0.01)
plt.plot(x, pol1(x),label=r'$l = 20 см$')


x2 = [46.7, 52.2, 56.2, 61.2, 70.9, 81.2, 93.8, 99.9, 85.1, 73.0, 64.3, 58, 52.7, 47.8]
y2 = [150, 165, 177.5, 245, 225, 257.5, 295, 315, 270, 230, 202.5, 185, 165, 150]
p2, v2 = np.polyfit(x2, y2, deg=1, cov=True)
plt.errorbar(x2, y2, xerr=4.6, yerr=14.7, fmt='.', ecolor='green')
x2 = np.arange(35, 110, 0.01)
plt.plot(x2, p2[0]*x2+p2[1], label=r'$l = 30 см$')


x3 = [38.8, 42.3, 45.5, 50.5, 67.6, 74.3, 83.6, 94.6, 79.3, 67.7, 60, 56.2, 53.3, 49]
y3 = [202.5, 220, 237.5, 265, 350, 387.5, 435, 492.5, 415, 350, 312.5, 295, 277.5, 255]
p3, v3 = np.polyfit(x3, y3, deg=1, cov=True)
plt.errorbar(x3, y3, xerr=4.5, yerr=23.4, fmt='.', ecolor='blue')
x3 = np.arange(35, 100, 0.01)
plt.plot(x3, p3[0]*x3+p3[1], label=r'$l = 50 см$')


plt.legend(loc='best', fontsize=12)
plt.xlabel(r'$I$, мА')
plt.ylabel(r'$U$, мВ')
plt.grid(True)
plt.tick_params("both", direction='in')
# plt.savefig('l.pdf')
plt.show()
```
